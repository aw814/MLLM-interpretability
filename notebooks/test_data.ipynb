Next, we check the `content` col to double-check if the text in the dataset entries match the reported language.
- Result: the content matches the tags and the reported distribution.

```python
import json
import pandas as pd
from collections import Counter
from langdetect import detect, DetectorFactory
from tqdm import tqdm

# For reproducibility
DetectorFactory.seed = 0

# --- Load your data ---
path = "../data/raw/eclektic_main.jsonl"

records = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        try:
            records.append(json.loads(line))
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(records)
print(f"âœ… Loaded {len(df)} records with {len(df.columns)} columns")

# --- Identify all content columns ---
content_cols = ['content']
print("\nğŸ§© Content columns detected:", content_cols)

# --- Detect language distribution for each *_c column ---
def safe_detect(text):
    try:
        return detect(text[:4000])
    except Exception:
        return "unknown"

lang_distributions = {}
detected_langs = []
for col in tqdm(content_cols, desc="Detecting languages per column"):
    sample_texts = df[col].dropna()
    detected = [safe_detect(t) for t in sample_texts]
    counts = Counter(detected)
    lang_distributions[col] = dict(counts)
    detected_langs.append(detected)

# --- Summarize results ---
print(f"\nLanguage distributions per column (based on col {content_cols}):")
for col, dist in lang_distributions.items():
    print(f"\n[{col}]")
    for lang, count in dist.items():
        print(f"  {lang}: {count}")

df["detected_lang_content"] = detected_langs[0]
```

Summary of QA entries:

```python
# Identify languages from *_q columns
qa_langs = sorted([c.split("_")[0] for c in df.columns if c.endswith("_q")])
summary = []

for lang in qa_langs:
    q_col = f"{lang}_q"
    if q_col not in df.columns:
        continue

    total_pairs = df[q_col].notna().sum()
    original_pairs = (df["original_lang"] == lang).sum()
    translated_pairs = total_pairs - original_pairs

    summary.append({
        "Language": lang,
        "Total_QA_Pairs": total_pairs,
        "Original": original_pairs,
        "Translated": translated_pairs,
        "Share_Original(%)": round(original_pairs / total_pairs * 100, 1) if total_pairs > 0 else 0
    })

qa_summary_df = pd.DataFrame(summary).sort_values("Total_QA_Pairs", ascending=False)

# --- Language code to full name mapping ---
lang_name_map = {
    "en": "English",
    "fr": "French",
    "de": "German",
    "he": "Hebrew",
    "hi": "Hindi",
    "id": "Indonesian",
    "it": "Italian",
    "ja": "Japanese",
    "ko": "Korean",
    "pt": "Portuguese",
    "es": "Spanish",
    "zh": "Chinese (Simplified)"
}

# --- Add full names to your summary DataFrame ---
qa_summary_df["Language_FullName"] = qa_summary_df["Language"].map(lang_name_map)

# Reorder columns for readability
qa_summary_df = qa_summary_df[
    ["Language", "Language_FullName", "Total_QA_Pairs", "Original", "Translated", "Share_Original(%)"]
]

qa_summary_df
```
